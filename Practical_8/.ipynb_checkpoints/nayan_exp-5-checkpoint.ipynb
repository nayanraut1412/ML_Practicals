# With Iris Dataset

```python
df = pd.read_csv("play-tennis.csv")
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python

df['outlook'] = encoder.fit_transform(df['outlook'])

df['temperature'] = encoder.fit_transform(df['temperature'])

df['humidity'] = encoder.fit_transform(df['humidity'])

df['wind'] = encoder.fit_transform(df['wind'])
```

```python
df.head
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
X = df.iloc[:,0:4]
Y = df.iloc[:,-1]
```

```python
X
```

```python
Y
```

```python
Y.shape
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=10)
```

```python
Y_test.shape
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
df.describe()
```

```python
from sklearn.naive_bayes import GaussianNB

dt=GaussianNB()
```

```python
dt.fit(X_train,Y_train)
```

```python
Y_pred = dt.predict(X_test)
```

```python
dt.score(X_test,Y_test)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score,f1_score,recall_score,precision_score
```

```python
print('Accuracy: %.3f' % accuracy_score(Y_test, Y_pred))
print('f1 score: %.3f' % f1_score(Y_test, Y_pred,average='micro'))
print('recall: %.3f' % recall_score(Y_test, Y_pred, average='macro'))
print('Precision: %.3f' % precision_score(Y_test, Y_pred,average='macro'))
```

```python
predicted_classes = dt.predict(X_test)
predicted_classes
```
