# 8) Aim: Train the Neural Network on a given datatset

```python
import tensorflow 
from tensorflow import keras 
from tensorflow.keras import Sequential 
from tensorflow.keras.layers import Dense, Flatten 
```

```python
(X_train,y_train),(X_test,y_test) = keras.datasets.mnist.load_data()
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.imshow(X_train[0]) 
```

```python
X_train = X_train/255 
X_test = X_test/255 
```

```python
model = Sequential() 
model.add(Flatten(input_shape=(28,28))) 
model.add(Dense(128, activation='relu')) 
model.add(Dense(32, activation='relu')) 
model.add(Dense (10, activation='softmax')) 
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='Adam', metrics=['accuracy'])       #remove unnessary columns means (reduce dimensions)
history = model.fit(X_train,y_train, epochs=25, validation_split=0.2)  #train test split
```

```python
y_prob= model.predict(X_test)
```

```python
y_pred= y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```

```python
plt.imshow(X_test[1])
```

```python
model.predict(X_test[1].reshape(1,28,28)).argmax(axis=1)
```

```python
(X_train,y_train),(X_test,y_test) = keras.datasets.fashion_mnist.load_data()
```

```python
X_test.shape
```

```python
X_train.shape
```

```python
y_train
```

```python
plt.imshow(X_train[450]) 
```

##### X_train = X_train/255 
X_test = X_test/255 

```python
model = Sequential() 
model.add(Flatten(input_shape=(28,28))) 
model.add(Dense(128, activation='relu')) 
model.add(Dense(32, activation='relu')) 
model.add(Dense (10, activation='softmax')) 
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='Adam', metrics=['accuracy'])       #remove unnessary columns means (reduce dimensions)
history = model.fit(X_train,y_train, epochs=40, validation_split=0.2)  #train test split
```

```python
y_prob= model.predict(X_test)
```

```python
y_pred= y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print("Test Accuracy:", accuracy)
print("Test loss:", loss)
```

```python
plt.plot(history.history['loss'])    #blue
plt.plot(history.history['val_loss'])   #orange
```

```python
plt.plot(history.history['accuracy'])  #blue
plt.plot(history.history['val_accuracy'])   #orange
```
