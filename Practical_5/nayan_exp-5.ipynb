# With Play-tennis Dataset

```python
df = pd.read_csv("play-tennis.csv")
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
df['outlook'] = encoder.fit_transform(df['outlook'])
df['temperature'] = encoder.fit_transform(df['temperature'])
df['humidity'] = encoder.fit_transform(df['humidity'])
df['wind'] = encoder.fit_transform(df['wind'])
```

```python
df.head
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
X = df.iloc[:,0:4]
Y = df.iloc[:,-1]
```

```python
X
```

```python
Y.shape
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=10)
```

```python
Y_test.shape
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
df.describe()
```

```python
from sklearn.naive_bayes import GaussianNB
dt=GaussianNB()
```

```python
dt.fit(X_train,Y_train)
```

```python
Y_pred = dt.predict(X_test)
```

```python
dt.score(X_test,Y_test)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score,f1_score,recall_score,precision_score
```

```python
print('Accuracy: %.3f' % accuracy_score(Y_test, Y_pred))
print('f1 score: %.3f' % f1_score(Y_test, Y_pred,average='micro'))
print('recall: %.3f' % recall_score(Y_test, Y_pred, average='macro'))
print('Precision: %.3f' % precision_score(Y_test, Y_pred,average='macro'))
```

```python
predicted_classes = dt.predict(X_test)
predicted_classes
```

# With Vehicle Dataset

```python
df1 = pd.read_csv("vehicle.csv")
```

```python
df1
```

```python
df1.isnull().sum()
```

```python
df1.info()
```

```python
df1.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python

df1['number_plate'] = encoder.fit_transform(df1['number_plate'])

df1['brand'] = encoder.fit_transform(df1['brand'])

df1['color'] = encoder.fit_transform(df1['color'])

df1['time'] = encoder.fit_transform(df1['time'])
```

```python
df1
```

```python
X = df1.iloc[:,1:4]
Y = df1.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=45)
```

```python
from sklearn.naive_bayes import GaussianNB
dt=GaussianNB()
```

```python
dt.fit(X_train,Y_train)
```

```python
Y_pred = dt.predict(X_test)
```

```python
dt.score(X_test,Y_test)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score,f1_score,recall_score,precision_score
```

```python
print('Accuracy: %.3f' % accuracy_score(Y_test, Y_pred))
print('f1 score: %.3f' % f1_score(Y_test, Y_pred,average='micro'))
print('recall: %.3f' % recall_score(Y_test, Y_pred, average='macro'))
print('Precision: %.3f' % precision_score(Y_test, Y_pred,average='macro'))
```

```python
predicted_classes = dt.predict(X_test)
predicted_classes
```
