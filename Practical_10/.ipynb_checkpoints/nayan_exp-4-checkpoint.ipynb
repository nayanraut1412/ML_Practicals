#   Aim: Build a decision tree model for a given dataset

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

# With Iris Dataset

```python
df = pd.read_csv("Iris.csv")
```

```python
df
```

```python
df.shape
```

```python
X=df.drop(columns='Species',axis=1)
```

```python
from sklearn.cluster import KMeans
km= KMeans(n_clusters=3,random_state=1)
km.fit(X)
```

```python
km.labels_
```

```python
km.cluster_centers_
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X['PetalLengthCm'],X['PetalWidthCm'],c=km.labels_)
plt.show()
```

```python
plt.scatter(X['SepalLengthCm'],X['SepalWidthCm'],c=km.labels_)
plt.show()
```

```python
inertia_scores=[]
for i in range(1,21):
    k=KMeans(n_clusters=i)
    k.fit(X)
    inertia_scores.append(k.inertia_)
```

```python
inertia_scores
```

```python
x_labels = range(1,21)
plt.plot(x_labels,inertia_scores,marker='o')
plt.show()
```
