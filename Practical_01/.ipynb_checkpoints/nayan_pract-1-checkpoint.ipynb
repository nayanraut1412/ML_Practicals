# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df1.head()
```

```python
encoder = LabelEncoder()
```

```python
new_car_data=pd.DataFrame()
new_car_data
```

```python
new_car_data['weight'] = encoder.fit_transform(df1['weight']) 
```

```python
new_car_data
```

```python
new_car_data['displacement'] = df1['displacement']
new_car_data['horsepower'] = df1['horsepower']
new_car_data['acceleration'] = encoder.fit_transform(df1['acceleration']) 
new_car_data['cylinders'] = df1['cylinders']
```

```python
new_car_data
```

```python
# Now all column data is in numner now we can handle missing data.
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
car_NaN_handle = imputer.fit_transform(new_car_data)
```

```python
car_NaN_handle
```

```python
car_NaN_handle=pd.DataFrame(car_NaN_handle)
```

```python
car_NaN_handle
```

```python
car_NaN_handle.isnull().sum()
```
